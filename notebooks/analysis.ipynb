#### 3. Volume vs. Margin Trade-Off
Let's analyse which category products generate high profits with lower margin or opposite...

2. Reduce Dependence on Low-Value Segment.

```python
# Create figure
fig, ax = plt.subplots(figsize=(14, 6))

# Calculate average margin
avg_margin = revenue_trends['pct_profit_margin'].mean()

# Plot line for profit margin
ax.plot(revenue_trends['order_month'], revenue_trends['pct_profit_margin'], 
        marker='o', linewidth=2.5, color='#DC143C', markersize=7, label='Profit Margin (%)')

# Add average margin line
ax.axhline(y=avg_margin, color='gray', linestyle='--', linewidth=2, alpha=0.7,
           label=f'Average: {avg_margin:.2f}%')

# Formatting
ax.set_title('Figure 7. Monthly Profit Margin Trends', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Profit Margin (%)', fontsize=12)
ax.legend(fontsize=13, loc='best')
ax.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()

# Save image
plt.savefig('../images/profit_margin_trends.png', dpi=300, bbox_inches='tight')
print("Profit Margin Trends chart saved as 'profit_margin_trends.png'")

# Display plot
plt.show()
```

**Key Insight:** Figure 7 reveals our profit margin evolution across 6+ years of operations, 
showing distinct phases:
1. Launch Volatility (2019).
- High variance: 48% to 57%.
- Small order volumes (10-185 orders/month).
- Normal startup behavior: margins swing due to sample size.
- Outcome: By Q4 2019, margins stabilized around 52% as volume increased.
2. Margin Stabilization (2020-2025).
- Remarkable consistency achieved an average margin of 51.91%.
- Remained in a range between 50.8% and 52.7% (only 1% variation).
- Maintained across 600x growth: From 203 orders/month (Jan 2020) to 6,819 orders/month (Nov 2025).
- Revenue scale: From $17K/month to $595K/month - margins held constant.

**Business Implications:**
* After stabilization in 2020, margins have varied by <1 percentage point for 5+ consecutive years. This may occur because:
    - Customers buy based on the product's or brand's value, not just the price
    - There is room to maintain prices even in competitive markets.
* Monthly revenue grew from $17K to $595K (+3,400%), while maintaining 52% profit margin. This may be the result of:
    - Variable costs being perfectly aligned with revenue.
    - Favorable supplier relationships in terms of volume.
    - Significant control over logistics and fulfillment costs.
* There was a slight drop to 50.84% (August 2021), but the recovery was immediate. This may indicate:
    - Rapid identification and correction of profit margin issues.
    - Strong financial controls were implemented.
    - Management has focused on profitability, not just revenue.

**Strategic Recommendations:**
1. Maintain Current Margin Strategy (51-52% Target) because:
    - Proven sustainability: 5+ years of consistency.
    - Suitable for the sector: E-commerce margins are typically 30-60%.
